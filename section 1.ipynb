```python
#given student marks dataset:
#calculate mean, mode, median.
#create a bar chart to visualize results
import pandas as pd
import numpy as np
student_marks={
    "student":["anna" , "brian", "grace" ,"catthy" , "kimo"],
    "math_score":["23" ,"54" ,"67" , "78" ,"98"],
}
df=pd.DataFrame( student_marks) 
df
```

```python
import numpy as np
math_score = np.array([ 23, 54 , 67 , 78, 98])
mean_mark= np.mean(math_score) 
print(mean_mark)               # the mean (avarage)
```

```python
import numpy as np
math_score = np.array([ 23, 54 , 67 , 78, 98])
median_mark= np.median(math_score) 
print(median_mark)            #the median (the middle value)
```

```python
import numpy as np
math_score = np.array([ 23, 54 , 67 , 78, 98])
maximum_mark=np.max(math_score)
print(maximum_mark)              # the highest
```

```python
import numpy as np
math_score = np.array([ 23, 54 , 67 , 78, 98])
minimum_mark=np.min(math_score)
print(minimum_mark)          # the lowest
```

```python
import matplotlib.pyplot as plt
student=["anna" , "brian", "grace" ,"catthy" , "kimo"]
math_score:["23" ,"54" ,"67" , "78" ,"98"]
plt.bar(student , math_score , color='lightgreen')
plt.xlabel('students')
plt.ylabel('math_score')
plt.title('student_marks')                     #the bar chart
plt.show()
```
