```python
import numpy as np
math_score = np.array([ 23, 54 , 67 , 78, 98])
mean_mark= np.mean(math_score) 
print(mean_mark)               # the mean (avarage)
```

```python
import numpy as np
math_score = np.array([ 23, 54 , 67 , 78, 98])
median_mark= np.median(math_score) 
print(median_mark)            #the median (the middle value)
```

```python
import numpy as np
math_score = np.array([ 23, 54 , 67 , 78, 98])
maximum_mark=np.max(math_score)
print(maximum_mark)              # the highest
```

```python
import numpy as np
math_score = np.array([ 23, 54 , 67 , 78, 98])
minimum_mark=np.min(math_score)
print(minimum_mark)          # the lowest
```

```python
import pandas as pd 
grouped=df.groupby(['guardian' , 'reason']).size()
print(grouped)
```

```python
grouped.columns =['guardian' , 'reason' , 'count']
print(grouped.columns)
```

```python
grouped=(
    df.groupby(['guardian', 'reason',])
       .size()
       .reset_index(name='count')
)
```

```python
import plotly.express as px
fig = px.scatter(
    grouped,
    x='guardian',
    y='reason',
    title='bubble chart',
    size='count',
    color='count',
    hover_name='count',
    size_max=60
)
fig.show()
```

```python
df['sex'] .value_counts().plot(kind='bar')                   # a single category
plt.title('number of student by sex')
plt.xlabel('sex')
plt.ylabel('number of students')
plt.show
```

```python
df['Mjob'] .value_counts().plot(kind='bar')                 # a single category
plt.title('distribution of mothers job')
plt.xlabel('Mjob')
plt.ylabel('number of students')
plt.show
```

```python
df['sex'] .value_counts().plot(kind='bar')        # abar chart for two categories (grouped bar chart)
cross_tab= pd.crosstab(df['sex'], df['guardian'])
cross_tab.plot(kind='bar')
plt.title('sex vs guardian')
plt.xlabel('sex')
plt.ylabel('number of students')
plt.xticks(rotation=0)                # shows how many males have mother as guardian and vice verse
plt.show
```
